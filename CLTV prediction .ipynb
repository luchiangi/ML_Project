# Customer life time value prediction

The objective of the project is to predict the CLTV based on the information collected from an IBM  customer database. The purpose of the project is to show how a company can  reduce its marketing cost by focusing on the most valuable customers. 

```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
#we took the data from https://community.ibm.com/accelerators/?context=analytics&query=tel&type=Data
#we have three dataframe: Telco_customer_churn_demographics,Telco_customer_churn,Telco_customer_churn_services)
#we need to merge them. 
```

```python
# Import the datafram on the customer demographics 

dfd = pd.read_excel('Telco_customer_churn_demographics.xlsx')
```

```python
# Import the datafram on the customer churn

dfc = pd.read_excel('Telco_customer_churn.xlsx')
```

```python
# Import the datafram on the customer churn services 

dfs = pd.read_excel('Telco_customer_churn_services.xlsx')
```

```python
dfs = dfs.rename(columns={'Customer ID':'CustomerID'})
```

```python
# Merge the dataframe of the customer churn with the dataframe of the the customer churn services 

dfcs = pd.merge(left=dfc, right=dfs, left_on='CustomerID', right_on='CustomerID')
```

```python
# Merge the previous dataframe merged with the dataframe of the the customer demographics 

df = pd.merge(left=dfcs, right=dfd, left_on='CustomerID', right_on='Customer ID')
```
